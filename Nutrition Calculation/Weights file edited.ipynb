```python
# IMporting libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Weight file being imported

WE = pd.read_csv('WEIGHT.csv')
```

```python
WE.head()
```

```python

# cleaning measurement method using lemmatizer and regex

import nltk
import re
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(text):
    
    s =""
    l = [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]
    
    for x in l:
        s +=" " + x
        break      # only first word is being taken
        
    s = re.sub('[^A-Za-z ]', '', s) # Only letters should remain here
    s = s.strip()
    return s

WE['measure_desc'] = WE.measure_desc.apply(lemmatize_text)

# replacing commas by space in WEIGHT table

WE["measure_desc"] = WE["measure_desc"].str.replace(","," ")

# changing some names to short forms

WE.loc[WE["measure_desc"] == "teaspoon", 'measure_desc'] = "tsp"
WE.loc[WE["measure_desc"] == "tablespoon", 'measure_desc'] = "tbsp"
WE.loc[WE["measure_desc"] == "fluid ounce", 'measure_desc'] = "fl oz"
WE.loc[WE["measure_desc"] == "ounce", 'measure_desc'] = "oz"
WE.loc[WE["measure_desc"] == "pound", 'measure_desc'] = "lb"
```

```python
WE.head()
```

```python
# Selecting relevent columns from WEIGHT table and dropping rest

WE =WE[['ndb_no','measure_desc','amount','grams']]
```

```python
# New column being created to help in calculations later on 

WE['grams_per_amount']=WE.grams/WE.amount
```

```python
# Just making sure ndb numbers are integers

WE["ndb_no"] = WE["ndb_no"].astype(int)
```

```python
# Creating a list of ndb numbers for which we need to define new units

ndblist = [x for x in WE["ndb_no"].values]
ndblist = set(ndblist)
ndblist =list(ndblist)
```

```python
# The measurement conversion file is being imported here

mea1 = pd.read_csv("Measurement_conversion - Sheet1.csv")
```

```python
mea1
```

```python
# Creating a list of possible conversions

conversion_units_available= list(mea1.columns)
```

```python
conversion_units_available
```

```python
# To clear display as we print count, to check progress of code

from IPython.display import clear_output
count =1

'''
for every ndbnumber create a list of measure_desc and make a standard from available units. If we can't make a standard then don't do anything
if we can get a standard then define rest of the units using that standard
'''
for i in ndblist: 
    count+=1
    clear_output(wait=True)
    print(count)
    ll = list(WE.loc[WE['ndb_no']==i]['measure_desc'])
    standard = 'none'                                   # default standard
    for item in ll:
        if item in conversion_units_available:
            standard = item
            standard_pergram = list(WE.loc[(WE['ndb_no']==i) & (WE['measure_desc']==standard)]['grams_per_amount'].values)
            standard_pergram=standard_pergram[0]
            break
    if standard =='none':
        pass
    else:
        for x in conversion_units_available:
            if x not in ll:
                gms = list((mea1[standard]/mea1[x]*standard_pergram).values)
                gms=gms[0]
                WE.loc[WE.index.max() + 1] =[i,x,1,gms,gms]



```
